```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

           ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

           ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

           ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

           ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
dataset
```

```python
model =TransactionEncoder()
model_m=model.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(model_m, columns=model.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemese = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemese
```

```python
from mlxtend.frequent_patterns import association_rules 
Rus = association_rules(frequent_itemese,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
Rus
```

```python
#RES1 = association_rules(frequent_itemese,metric="lift",min_threshold=1.25)
#RES1
```

```python
Res1 = Rus[['antecedent support','consequent support','support','confidence','lift']]
Res1
```

```python
RES2 = Res1[Res1['confidence']>=1]
RES2
```

```python
import csv
bigdata = []
with open('Market_Basket_Optimisation.csv', newline='') as f:
    reader = csv.reader(f)
    for b in reader:
         bigdata.append(b)
```

```python
model1 =TransactionEncoder()
model1_m=model1.fit(bigdata).transform(bigdata)    #Apply one-hot-encoding on our dataset
df2=pd.DataFrame(model1_m, columns=model1.columns_)  #Creating a new DataFrame from our Numpy array
df2

```

```python
df2.columns
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemese = apriori(df2, min_support=0.005, use_colnames=True)
frequent_itemese
```

```python
from mlxtend.frequent_patterns import association_rules 
RUS = association_rules(frequent_itemese,metric="confidence",min_threshold=0.5) # associate itemsets with confidence over 70%.
RUS
```

```python
RUS1= RUS[['antecedent support','consequent support','support','confidence','lift']]
RUS1
```
